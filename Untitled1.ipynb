```python
# ✅ 셀 1: 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# ✅ 셀 2: 정상 트래픽 + DDoS 트래픽 CSV 불러오기
df_normal = pd.read_csv("MachineLearningCVE/Monday-WorkingHours.pcap_ISCX.csv")
df_ddos = pd.read_csv("MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")

print("정상:", df_normal.shape)
print("DDoS:", df_ddos.shape)
```

```python
# ✅ 셀 3: 라벨링 추가 후 병합 (BENIGN → 0, DDoS → 1)
df_normal["Label"] = "BENIGN"
df_ddos["Label"] = "DDoS"
df = pd.concat([df_normal, df_ddos], ignore_index=True)
print("전체:", df.shape)
print(df.columns.tolist())
```

```python
# ✅ 셀 4: Label 컬럼 클린하게 정리하고 다시 삽입
df.columns = df.columns.str.strip()  # 공백 제거

# 1. 'Label' 이름 가진 모든 컬럼 삭제
df = df.loc[:, ~df.columns.str.strip().str.lower().isin(['label'])]

# 2. 정답값 다시 삽입
df['Label'] = [0] * len(df_normal) + [1] * len(df_ddos)

# 3. 확인
print("컬럼 수:", len(df.columns))
print("Label 분포:\n", df["Label"].value_counts())
```

```python
# ✅ 셀 5: Label을 X에서 제거하고 숫자형 피처만 남기기
y = df["Label"]
X = df.drop(columns=["Label"], errors='ignore')
X = X.select_dtypes(include=["number"])

# 확인
print("'Label' in X.columns? →", 'Label' in X.columns)  # 반드시 False
print("X shape:", X.shape)
# ✅ 무한값과 NaN 제거
X = X.replace([np.inf, -np.inf], np.nan)  # 무한값 → NaN으로 변환
X = X.dropna()  # NaN 제거
y = y[X.index]  # X에서 행이 지워졌으므로 y도 인덱스 맞춰 재정렬
```

```python
# ✅ 셀 6: 학습/테스트 데이터셋 나누기
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# ✅ 셀 7: RandomForestClassifier 학습
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)
```

```python
# ✅ 셀 8: 모델 예측 및 성능 평가
y_pred = model.predict(X_test)

print("정확도 (Accuracy):", accuracy_score(y_test, y_pred))
print("정밀도 (Precision):", precision_score(y_test, y_pred))
print("재현율 (Recall):", recall_score(y_test, y_pred))
print("F1 점수 (F1-score):", f1_score(y_test, y_pred))

print("\n📋 분류 리포트:")
print(classification_report(y_test, y_pred, target_names=['BENIGN', 'DDoS']))
```

```python
# ✅ 셀 9: Confusion Matrix 시각화
# ✅ 한글 폰트 설정 (윈도우용)
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 경로 찾기 (예: 'Malgun Gothic')
font_path = "C:/Windows/Fonts/malgun.ttf"  # 또는 나눔고딕 사용 가능

# 폰트 등록
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)

# 마이너스 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

print(f"적용된 폰트: {font_name}")
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['BENIGN', 'DDoS'], yticklabels=['BENIGN', 'DDoS'])
plt.xlabel("예측값")
plt.ylabel("실제값")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
# ✅ 셀 10: 상위 15개 중요 피처 시각화
importances = model.feature_importances_
feature_names = X.columns
sorted_idx = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
sns.barplot(x=importances[sorted_idx][:15], y=feature_names[sorted_idx][:15])
plt.title("Top 15 Feature Importances")
plt.xlabel("중요도 점수")
plt.ylabel("피처")
plt.tight_layout()
plt.show()
```
