```python
# ‚úÖ ÏÖÄ 1: ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# ‚úÖ ÏÖÄ 2: Ï†ïÏÉÅ Ìä∏ÎûòÌîΩ + DDoS Ìä∏ÎûòÌîΩ CSV Î∂àÎü¨Ïò§Í∏∞
df_normal = pd.read_csv("MachineLearningCVE/Monday-WorkingHours.pcap_ISCX.csv")
df_ddos = pd.read_csv("MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")

print("Ï†ïÏÉÅ:", df_normal.shape)
print("DDoS:", df_ddos.shape)
```

```python
# ‚úÖ ÏÖÄ 3: ÎùºÎ≤®ÎßÅ Ï∂îÍ∞Ä ÌõÑ Î≥ëÌï© (BENIGN ‚Üí 0, DDoS ‚Üí 1)
df_normal["Label"] = "BENIGN"
df_ddos["Label"] = "DDoS"
df = pd.concat([df_normal, df_ddos], ignore_index=True)
print("Ï†ÑÏ≤¥:", df.shape)
print(df.columns.tolist())
```

```python
# ‚úÖ ÏÖÄ 4: Label Ïª¨Îüº ÌÅ¥Î¶∞ÌïòÍ≤å Ï†ïÎ¶¨ÌïòÍ≥† Îã§Ïãú ÏÇΩÏûÖ
df.columns = df.columns.str.strip()  # Í≥µÎ∞± Ï†úÍ±∞

# 1. 'Label' Ïù¥Î¶Ñ Í∞ÄÏßÑ Î™®Îì† Ïª¨Îüº ÏÇ≠Ï†ú
df = df.loc[:, ~df.columns.str.strip().str.lower().isin(['label'])]

# 2. Ï†ïÎãµÍ∞í Îã§Ïãú ÏÇΩÏûÖ
df['Label'] = [0] * len(df_normal) + [1] * len(df_ddos)

# 3. ÌôïÏù∏
print("Ïª¨Îüº Ïàò:", len(df.columns))
print("Label Î∂ÑÌè¨:\n", df["Label"].value_counts())
```

```python
# ‚úÖ ÏÖÄ 5: LabelÏùÑ XÏóêÏÑú Ï†úÍ±∞ÌïòÍ≥† Ïà´ÏûêÌòï ÌîºÏ≤òÎßå ÎÇ®Í∏∞Í∏∞
y = df["Label"]
X = df.drop(columns=["Label"], errors='ignore')
X = X.select_dtypes(include=["number"])

# ÌôïÏù∏
print("'Label' in X.columns? ‚Üí", 'Label' in X.columns)  # Î∞òÎìúÏãú False
print("X shape:", X.shape)
# ‚úÖ Î¨¥ÌïúÍ∞íÍ≥º NaN Ï†úÍ±∞
X = X.replace([np.inf, -np.inf], np.nan)  # Î¨¥ÌïúÍ∞í ‚Üí NaNÏúºÎ°ú Î≥ÄÌôò
X = X.dropna()  # NaN Ï†úÍ±∞
y = y[X.index]  # XÏóêÏÑú ÌñâÏù¥ ÏßÄÏõåÏ°åÏúºÎØÄÎ°ú yÎèÑ Ïù∏Îç±Ïä§ ÎßûÏ∂∞ Ïû¨Ï†ïÎ†¨
```

```python
# ‚úÖ ÏÖÄ 6: ÌïôÏäµ/ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ÏÖã ÎÇòÎàÑÍ∏∞
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# ‚úÖ ÏÖÄ 7: RandomForestClassifier ÌïôÏäµ
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)
```

```python
# ‚úÖ ÏÖÄ 8: Î™®Îç∏ ÏòàÏ∏° Î∞è ÏÑ±Îä• ÌèâÍ∞Ä
y_pred = model.predict(X_test)

print("Ï†ïÌôïÎèÑ (Accuracy):", accuracy_score(y_test, y_pred))
print("Ï†ïÎ∞ÄÎèÑ (Precision):", precision_score(y_test, y_pred))
print("Ïû¨ÌòÑÏú® (Recall):", recall_score(y_test, y_pred))
print("F1 Ï†êÏàò (F1-score):", f1_score(y_test, y_pred))

print("\nüìã Î∂ÑÎ•ò Î¶¨Ìè¨Ìä∏:")
print(classification_report(y_test, y_pred, target_names=['BENIGN', 'DDoS']))
```

```python
# ‚úÖ ÏÖÄ 9: Confusion Matrix ÏãúÍ∞ÅÌôî
# ‚úÖ ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (ÏúàÎèÑÏö∞Ïö©)
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# ÌïúÍ∏Ä Ìè∞Ìä∏ Í≤ΩÎ°ú Ï∞æÍ∏∞ (Ïòà: 'Malgun Gothic')
font_path = "C:/Windows/Fonts/malgun.ttf"  # ÎòêÎäî ÎÇòÎàîÍ≥†Îîï ÏÇ¨Ïö© Í∞ÄÎä•

# Ìè∞Ìä∏ Îì±Î°ù
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)

# ÎßàÏù¥ÎÑàÏä§ Íπ®Ïßê Î∞©ÏßÄ
plt.rcParams['axes.unicode_minus'] = False

print(f"Ï†ÅÏö©Îêú Ìè∞Ìä∏: {font_name}")
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['BENIGN', 'DDoS'], yticklabels=['BENIGN', 'DDoS'])
plt.xlabel("ÏòàÏ∏°Í∞í")
plt.ylabel("Ïã§Ï†úÍ∞í")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
# ‚úÖ ÏÖÄ 10: ÏÉÅÏúÑ 15Í∞ú Ï§ëÏöî ÌîºÏ≤ò ÏãúÍ∞ÅÌôî
importances = model.feature_importances_
feature_names = X.columns
sorted_idx = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
sns.barplot(x=importances[sorted_idx][:15], y=feature_names[sorted_idx][:15])
plt.title("Top 15 Feature Importances")
plt.xlabel("Ï§ëÏöîÎèÑ Ï†êÏàò")
plt.ylabel("ÌîºÏ≤ò")
plt.tight_layout()
plt.show()
```
